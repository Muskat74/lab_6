```python
df = pd.read_excel('C:\\Users\\vsevo\\Desktop\\FOR_LAB_6.xlsx')
sns.set_palette('pastel')

df.head()
```

```python
# 1. Time series
plt.figure(figsize=(10,6))
sns.lineplot(data=df, x='Order Date', y='Sales')
plt.title('Time Series of Sales')
plt.show()
```

```python
#вартість доставки для кожного режиму доставки
shipping_costs = df.groupby('Ship Mode')['Shipping Cost'].sum().sort_values(ascending=False)

#діаграма "donut"
fig, ax = plt.subplots()
wedges, _ = ax.pie(shipping_costs, startangle=90, pctdistance=0.85, wedgeprops=dict(width=0.3))

#білий круг у центрі
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

ax.axis('equal')  
plt.tight_layout()

#значення до кожного сектора
bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
kw = dict(xycoords='data', textcoords='data', arrowprops=dict(arrowstyle="-"), zorder=0, va="center")

for i, p in enumerate(wedges):
    ang = (p.theta2 - p.theta1)/2. + p.theta1
    y = np.sin(np.deg2rad(ang))
    x = np.cos(np.deg2rad(ang))
    horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
    connectionstyle = "angle,angleA=0,angleB={}".format(ang)
    kw["arrowprops"].update({"connectionstyle": connectionstyle})
    value = int(shipping_costs[i])
    if value >= 1000:
        value_str = f'{value/1000}K'
    else:
        value_str = str(value)
    ax.annotate(shipping_costs.index[i] + ' $' + value_str, xy=(x, y), xytext=(1.35*np.sign(x), 1.4*y),
                 horizontalalignment=horizontalalignment, **kw)

plt.show()
```

```python
#Завантажую дані з аркуша 'Orders'
df_orders = pd.read_excel('C:\\Users\\vsevo\\Desktop\\FOR_LAB_6.xlsx', sheet_name='Orders')

#Завантажую дані з аркуша 'Customers'
df_customers = pd.read_excel('C:\\Users\\vsevo\\Desktop\\FOR_LAB_6.xlsx', sheet_name='Customers')

#Міняю 'Customer ID' на 'Customer Name'
df_orders = df_orders.merge(df_customers[['Customer ID', 'Customer Name']], on='Customer ID', how='left')

#Сортую дані за 'Sales' перші 10 рядків
df_orders = df_orders.sort_values('Sales', ascending=False).head(10)

#3. Sales by customer
plt.figure(figsize=(10,6))
barplot = sns.barplot(data=df_orders, y='Customer Name', x='Sales', orient='h')
plt.title('Sales by Top 10 Customers')

#Додаю кількість продажів до кінця кожного стовпця
for i, v in enumerate(df_orders['Sales']):
    plt.text(v + 3, i + .25, str(int(v/1000)) + 'K', color='black', fontweight='light')

#Міняюмаштаб осі X, щоб відображати тисячі як 'K'
ticks_loc = plt.xticks()[0]
plt.xticks(ticks_loc, [f'{int(x/1000)}K' if x != 0 else '0' for x in ticks_loc], rotation=0)

plt.show()
```

```python
df_orders['Delivery Time'] = (df_orders['Ship Date'] - df_orders['Order Date']).dt.days

df_delivery_time = df_orders.groupby('Ship Mode')['Delivery Time'].mean().reset_index()

plt.figure(figsize=(15,6)) 
sns.barplot(data=df_delivery_time, x='Ship Mode', y='Delivery Time', dodge=False)
plt.title('Average Delivery Time by Ship Mode')
plt.show()
```
